```python
import numpy as np
import pickle
```

```python
test_worm_1_pickle_path = "/home/daniel/pythonny/MPI/CElegans/worm_data/test1worms.pkl"
test_worm_2_pickle_path = "/home/daniel/pythonny/MPI/CElegans/worm_data/test2worms.pkl"
subgraph_output_directory = "/home/daniel/pythonny/MPI/CElegans/subgraph_testing_data"
```

```python
test_sizes = [i for i in range(10, 560, 10)]
test_sizes.append(558)
test_sizes = np.array(test_sizes)
```

```python
with open(test_worm_1_pickle_path, "rb") as f:
    worm1 = pickle.load(f)
with open(test_worm_2_pickle_path, "rb") as f:
    worm2 = pickle.load(f)
all_worms = worm1 + worm2
```

```python
num_samples = np.ceil(558 / test_sizes) * 2 # Could sample less by figuring out optimal number of samples to reduce noise in variance estimate but oversampling was easier
```

```python
rng = np.random.default_rng(42)

for subgraph_size in test_sizes:
    test_worms = []
    for worm in all_worms:
        answer, coords = worm
        for sample in range(np.ceil(558 / subgraph_size).astype(int)*2):
            idx = rng.choice(len(answer), size=min(subgraph_size, len(answer)), replace=False)
            test_worms.append([answer[idx], coords[idx]])
    with open(f"{subgraph_output_directory}/subgraph_{subgraph_size}.pkl", "wb") as f:
        pickle.dump(test_worms, f)
```
